# Deep Dive into LangChain
A practical and conceptual guide with working code examples.

## 1. Introduction to LangChain
LangChain is a framework for building applications powered by LLMs with orchestration, chaining, and tool integration.

### Basic LLM Call

```
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)

response = llm.invoke("Explain LangChain in one sentence.")
print(response.content)
```

### Prompt Template Example

```
from langchain.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms."
)

print(template.format(topic="LangChain"))
```

### Chain Example

```
from langchain_core.output_parsers import StrOutputParser

chain = template | llm | StrOutputParser()
print(chain.invoke({"topic": "LangChain"}))
```

### Memory Example

```
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

memory = ConversationBufferMemory()

conversation = ConversationChain(llm=llm, memory=memory)

print(conversation.predict(input="Hi, I am Alex"))
print(conversation.predict(input="What is my name?"))
```

### Agent with Tool

```
from langchain.agents import initialize_agent, Tool

def add(a: int, b: int):
    return a + b

tools = [
    Tool(
        name="Adder",
        func=lambda x: add(*map(int, x.split())),
        description="Adds two numbers"
    )
]

agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)

agent.run("Add 10 and 20")
```

### Vector Store Example

```
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()

vector_store = FAISS.from_texts(
    ["LangChain is powerful", "LLMs are transformative"],
    embeddings
)

results = vector_store.similarity_search("What is LangChain?")
print(results[0].page_content)
```
